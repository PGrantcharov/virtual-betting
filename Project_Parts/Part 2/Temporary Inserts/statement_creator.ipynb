```python
# Creates a dictionary with all the short form names for g_id
def get_short(team):
    mid = team.split()[-1]
    for team in game.home_name.unique():
        if mid in team:
            long = team
            break
    return game.loc[game.home_name == long].iloc[0].game_id[-3:]

shorts = {}
for team in bet_tl.home.unique():
    shorts[team] = get_short(team)

# functions that decodes the betting lines, returns as string tuple
def get_total(line):
    if '-' in line:
        odds_payout = '-' + line.split('-')[1]
        odds_line = line.split('-')[0].replace('Â½', '.5')
    else:
        odds_payout = '+' + line.split('+')[1]
        odds_line = line.split('+')[0].replace('Â½', '.5')
    return odds_payout, odds_line
def get_spread(line):
    if '-' in line[1:]:
        odds_payout = '-' + line[1:].split('-')[1]
        odds_line = line[0] + line[1:].split('-')[0].replace('Â½', '.5')
    else:
        odds_payout = '+' + line[1:].split('+')[1]
        odds_line = line[0] + line[1:].split('+')[0].replace('Â½', '.5')
    return odds_payout, odds_line

def odds_convert(line):
    line = int(line)
    if line < 0:
        return round(100/(line*-1) + 1, 3)
    else:
        return round((line/100) + 1, 3)
```

```python
# schema for make_odds table
# o_id | g_id | sb_id | bt_id | odds_time | odds_side | odds_payout | odds_line 
def get_gid(date, num):
    home = bet_tl.loc[(bet_tl.date == date) & (bet_tl.game_num == num)].home.iloc[0]
    return str(date) + '0' + shorts[home]

# has the bet id numbers for the database
btid = {'p':'3', 'm':'1', 't':'2'}

counter = 0
for index, el in bet.iterrows():
    if el.length != 'full':
        continue
    g_id = get_gid(el.date, el.game_num)
    bt_id = btid[el.bet]
    odds_time = game.loc[game.game_id == g_id].date.iloc[0]
    for sb in range(1, 11):                                      # loop 10 books
        sb_id = str(sb)                                           
        for side in ['ab', 'hb']:                                # loop V/H side
            name = side + str(sb)
            if el[name] == '-':                                  # skips if blank
                continue
                
            if el.bet == 't':                                    # if total line
                odds_side = "O" if side == 'ab' else 'U'
                odds_payout, odds_line = get_total(el[name])
            else:
                odds_side = "V" if side == 'ab' else 'H'
                if el.bet == 'p':                                # if spread line
                    odds_payout, odds_line = get_spread(el[name])
                else:                                            # if money line
                    odds_payout, odds_line = (el[name], '0')
            print("INSERT INTO make_odds VALUES(" + 
                  str(counter) + ",'" + g_id + "'," +
                  str(sb_id) + "," +
                  str(bt_id) + ",'" +
                  str(odds_time.replace(',', '')) + "','" +
                  str(odds_side) + "'," +
                  str(odds_convert(odds_payout)) + "," +
                  str(odds_line) + ");"
                 ) 
            counter += 1
```
